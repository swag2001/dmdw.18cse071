```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/swag2001/dmdw.18cse071/main/vgsales.csv", sep=',') 
df
```

```python
df.head()
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
plt.scatter(df['Year'],df['Global_Sales'],c='orange')
plt.xlabel('Years')
plt.ylabel('Global_Sales')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.hist(df['Genre'],color = 'blue',edgecolor = 'grey',bins = 12)
```

```python
plt.figure(figsize=(10,10))
plt.hist(df['Genre'],color = 'skyblue',edgecolor = 'grey',bins = 12)
plt.title('Histograms of Genres')
plt.xlabel('Genres')
plt.ylabel('Frequency')
plt.show()
```

```python
counts = [969, 120, 12, 97, 279]
Platform = ('Misc', 'GB', 'Wii','PS2','PS3')
index = np.arange(len(Platform))
plt.bar(index, counts, color=['orange', 'blue', 'red','violet','pink'])
plt.title('Bar plot of Platforms')
plt.xlabel('Platform Types')
plt.ylabel('Frequency')
plt.xticks(index, Platform, rotation = 90)
plt.show()
```
